```python
from astropy.table import Table
import pyvo


"""Get information of galaxies in the fornax cluster from SIMBAD"""
```

```python
# Make the TAP service object
service = pyvo.dal.TAPService ("http://simbad.u-strasbg.fr/simbad/sim-tap")
```

```python
# Query the TAP service with a simple ADQL query.
result = service.search ("""
SELECT TOP 15
main_id,oid,ra,dec,galdim_majaxis
FROM basic
WHERE 1=CONTAINS(
POINT('',ra, dec),
CIRCLE('',54.81426, -35.4652, 0.6))
AND galdim_majaxis>0.4 """)
print ( result )
```

```python
# Upload Your list of Names
a=['3C 10','M 2','HD 3']
table = Table([a], names=['myname'], meta={'meta':'first table'})
```

```python
from astroquery.utils.tap.core import TapPlus
service = TapPlus ("http://simbad.u-strasbg.fr/simbad/sim-tap")
```

```python
full_qualified_table_name = 'TAP_UPLOAD.my_table'

query = 'SELECT * FROM ' + full_qualified_table_name
job = service.launch_job(query=query,upload_resource=table, \
upload_table_name="my_table", verbose=True)

results = job.get_results()
print(results)
```

```python

query = 'SELECT myname, main_id AS "simbadName", rvz_redshift, rvz_bibcode FROM basic JOIN ident ON oid=oidref RIGHT JOIN ' + full_qualified_table_name + ' ON myname=id'
job = service.launch_job(query=query,upload_resource=table, \
upload_table_name="my_table")

results = job.get_results()
print(results)
```
